```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from time import time
import os


# 生成数据
X, y = make_classification(n_samples=1000, n_features=50, n_informative=15, n_redundant=5, random_state=0)

running_time = []
# 测试不同核心数随机森林运行时间
for n in range(os.cpu_count()):
    
    begin = time()
    model = RandomForestRegressor(n_estimators=1000, n_jobs=n+1)
    model.fit(X, y)
    end = time()
    running_time.append(end-begin)
plt.plot(range(1, os.cpu_count()+1), running_time)
plt.xlabel("n_jobs")
plt.ylabel("running_time")
plt.show()
```
